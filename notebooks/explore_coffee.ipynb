```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import data
df = pd.read_csv("../data/crop/coffee_output.csv")
```

```python
# TODO: view first 5 rows of your data
df.head()
```

```python
# TODO: describe your numeric columns
df.describe()
```

```python
# TODO: display the null values in your dataframe
df.isnull()

```

```python
# TODO: drop your null values
df.dropna()
```

```python
# TODO: Select only Minas Gerais

#df=['subdivision'] ==['Minas Gerais']
Minas_Gerais= df[df['subdivision']=='Minas Gerais']
Minas_Gerais.head()
```

```python
# TODO: BEGIN YOUR EDA!


ax =sns.lineplot(x='year', y='bearing_trees', data= Minas_Gerais, color ='green', marker ='o')
ax.set_ylabel('Bearing trees(millions)')
ax.set_title('Number of Bearing Trees By Year')

df.plot.hist(x='year', y='bearing_trees', bins=15)
plt.title('Distribution of Bearing Trees by Year')
plt.xlabel=('year')
plt.ylabel = ('Bearing trees')
```

```python
ax =sns.lineplot(x='year', y='bearing_hectares', data= Minas_Gerais, color ='red', marker ='o')
ax.set_ylabel('Bearing hectares(millions)')
ax.set_title('Number of Bearing Hectares By Year')

df.plot.hist(x='year', y='bearing_hectares', bins=15)
plt.title('Distribution of Bearing Hectares by Year')
plt.xlabel=('year')
plt.ylabel = ('Bearing Hectares')

df = pd.pivot_table(df, index =['year'], values =['bearing_hectares'] )
print(df)

```
