```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import coffee data
coffee = pd.read_csv("../data/crop/coffee_output.csv")

# import weather data
weather = pd.read_csv("../data/weather/weather_all.csv")
```

```python
# TODO: Merge these two dataframes together

analysis_all =pd.merge(weather, coffee , on= ['year', 'subdivision'],how='outer')

analysis_all.to_csv('../data/weather/analysis_all.csv', index=False)
```

```python
# TODO: Check for null values 
analysis_all.isnull()
```

```python
# TODO: Drop rows that contain null values
Analysis_clean = analysis_all.dropna()

Analysis_clean.head()
```
